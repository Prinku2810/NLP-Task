```python
dataset = pd.read_csv("/content/aspell.txt")
```

```python
dataset.columns=['incorrect']
```

```python
dt=dataset.head(100)
dt
```

```python

wrng = []
crt = []

for i in dt['incorrect']:
  if ':' in i:
    # Splitting the sentence on '|'
    str1 = i.split(':')
    # Previous word to '|' is storing in loriginal list.
    crt.append(str1[0])
    # Next word to '|' is storing in lcorrected list.
    wrng.append(str1[1])

wrng
```

```python
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from collections import Iterable
#def flatten(lis):
#     for item in lis:
#         if isinstance(item, Iterable) and not isinstance(item, str):
#             for x in flatten(item):
#                 yield x
#         else:        
#             yield item

token_list = []
for token in wrng:
    token_list.append(word_tokenize(token))
incorrect_words = [element for innerList in token_list for element in innerList]
#incorrect_words=list(flatten(token_list))
incorrect_words
```

```python
from nltk.metrics.distance import edit_distance
nltk.download('words')
from nltk.corpus import words
correct_words = words.words()


for word in incorrect_words:
    temp = [(edit_distance(word, w),w) for w in correct_words if w[0]==word[0]]
    print(sorted(temp, key = lambda val:val[0])[0][1])

```
